```
# importing all the basic libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# importing all the keras libraries.
from keras.utils import to_categorical
from tensorflow.python.keras.models import Sequential, Model
from tensorflow.python.keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Lambda, Input, Dense, Dropout, Flatten
from tensorflow.python.keras.layers.advanced_activations import LeakyReLU,ThresholdedReLU
from tensorflow.python.keras.layers.normalization import BatchNormalization
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.optimizers import SGD,Adam,Nadam,Adamax,TFOptimizer
from keras.utils import to_categorical
from tensorflow.python.keras.callbacks import TensorBoard

num_of_classes = 5    # storing number of classes in a variable
epochs = 50           # storing epoches in a variable # epoches = 100
shift_fraction=0.005

# defining the shape of the images
image_row = 28
image_column = 28
batch_size = 512
image_shape = (image_row,image_column,1)
```

```
# importing the datset
dataset_train = pd.read_csv("train.csv")
dataset_test = pd.read_csv("testX.csv")

# converting the dataset's data to float32 format because keras and tensorflow uses that format
dataset_train = np.array(dataset_train,dtype='float32')
dataset_test = np.array(dataset_test,dtype='float32')

# splitting the dataset into X and y (features and clasees) dividing them with 255 because we are rescaling them to 0 and 1
X_train = dataset_train[:,2:]/255
y_train = dataset_train[:,1]

X_test = dataset_test[:,1:]/255

# use for real output and to get real prediction
X_train = X_train.reshape(X_train.shape[0],image_row,image_column,1)
X_test = X_test.reshape(X_test.shape[0],image_row,image_column,1)

y_train = to_categorical(y_train,num_classes=num_of_classes)
```

```
# defining the cnn model
cnn_model = Sequential()

cnn_model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=image_shape))
cnn_model.add(BatchNormalization())
#cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(MaxPooling2D(pool_size=(2,2))) # downsampling the output from 28*28 to 14*14
cnn_model.add(Dropout(0.5))

cnn_model.add(Conv2D(filters=128,kernel_size=(3,3),activation='relu'))
cnn_model.add(BatchNormalization())
#cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(MaxPooling2D(pool_size=(2,2))) # downsampling the output from 28*28 to 14*14
cnn_model.add(Dropout(0.5))

cnn_model.add(Flatten())

cnn_model.add(Dense(512,activation='relu'))
cnn_model.add(Dropout(0.5))
#cnn_model.add(LeakyReLU(alpha=0.1))
cnn_model.add(Dense(num_of_classes,activation='softmax'))
cnn_model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
cnn_model.summary()
```

```
# Parameter for Image Data Augumentation
gen = ImageDataGenerator(width_shift_range=shift_fraction,height_shift_range=shift_fraction,horizontal_flip=True)
batches = gen.flow(X_train, y_train, batch_size=batch_size)
#steps_per_epoch=X_train.shape[0]//batch_size
fashion_train=cnn_model.fit_generator(batches, steps_per_epoch=1000, epochs=epochs)
```

```
output = cnn_model.predict(X_test)

# converting the output back to normal.
final_output = np.argmax(output,axis=-1)

# getting the csv file
Id_output = []
for i in range(len(final_output)):
  Id_output.append(i)

output_csv = pd.DataFrame({'Id':Id_output, 'Label':final_output})

output_csv.to_csv('best_CNN.csv',index=False)
```
